Question 3. Write a Python program to generate 12 random numbers between 1 and 100. Sort the list of numbers in
ascending order and print it out. Sort the list of numbers in descending order and print it out.
Calculate the average for the group and print it out.

```python
import random
random_num = random.sample(range(1, 100), 12)
print(' 12 random numbers fall between 1 and 100: \n ', random_num)
ascending = sorted(random_num)
print('ascending numbers are: \n' , ascending)
descending = sorted(random_num,reverse=True)
print('descending numbers are: \n' , descending)
print("the random numbers", ascending, 'the avrage is: ', sum(random_num)/len(random_num))
```
