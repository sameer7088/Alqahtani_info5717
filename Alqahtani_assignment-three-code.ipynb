```python
class Student:
  def __init__(self,first,middle,last):
    self.first_name = first
    self.middle_name = middle
    self.last_name = last
    self.euid = ""
    self.gpa = 0.00
    self.classes = []
    

def student_input():
    temp_first = input("Enter the student's first name: ")
    temp_middle = input("Enter the student's middle name: ")
    temp_last = input("Enter the student's last name: ")
    temp_student = Student(temp_first,temp_middle,temp_last)
    return temp_student

student_1 = student_input()
student_2 = student_input()
student_3 = student_input()

import random
def euid_generator(student):
    number = random.randint(1000,9999) 
    if student.middle_name:
        x = student.first_name[0].lower()
        y = student.middle_name[0].lower() 
        z = student.last_name[0].lower() 
        student.euid = x+y+z+str(number) 
        print(student.first_name+" "+student.middle_name+" "+student.last_name+" - "+student.euid)
    else: 
        x = student.first_name[0].lower() 
        z = student.last_name[0].lower() 
        student.euid = x+z+str(number) 
        print(student.first_name+" "+student.last_name+" - "+student.euid)
        
print("\n")
euid_generator(student_1)
euid_generator(student_2)
euid_generator(student_3)


def register(student):
    class_number = input("Enter the class number: ")
    semester = input("Enter the semester: ")
    grade = input('Enter the grade received: ')
    student.classes.append((class_number,semester,grade)) 
    

print("\nClass information for "+student_1.first_name+" "+student_1.last_name)
register(student_1)
register(student_1)

print("\nClass information for "+student_2.first_name+" "+student_2.last_name)
register(student_2)
register(student_2)

print("\nClass information for "+student_3.first_name+" "+student_3.last_name)
register(student_3)
register(student_3)

def print_registration(student):
    print("\n"+student.first_name+" "+student.last_name+" ("+student.euid+") is registered for:")
    for item in student.classes: 
        print(item[0]+"\t"+item[1]+"\t"+item[2])

print_registration(student_1)
print_registration(student_2)
print_registration(student_3)

def gpa_calculator(student):
    total = 0 
    for item in student.classes: 
        grade = item[2].lower() 
        if grade == "a": 
            total += 4
        elif grade == "b": 
            total += 3
        elif grade == "c": 
            total += 2
        elif grade == "d":
            total += 1
    
    student.gpa = round(total/len(student.classes),2) 
    print("\nThe GPA for "+student.first_name+" "+student.last_name+" is "+str(student.gpa))

gpa_calculator(student_1)
gpa_calculator(student_2)
gpa_calculator(student_3)
```

```python
def quiz_introduction():
   print("*" * 24)
   print("** A Simple Math Quiz **")
   print("*" * 24)

def quiz_questions_init():
   print("1. Addition")
   print("2. Subtraction")
   print("3. Multiplication")
   print("4. Integer Division")
   print("5. Exit")

def section_separator():

  print("-" * 24)

def get_user_question_choice():
    val = int(input("Enter your choice: "))
    return val

def quiz_questions(option, count):
  import random
  a1 = random.randint(1,99)
  a2 = random.randint(1,99)
  if option == 1:
     problem = str(a1) + "+" + str(a2)
     answer = a1 + a2
  elif option == 2:
     problem = str(a1) + "-" + str(a2)
     answer = a1 - a2
  elif option == 3:
     problem = str(a1) + "*" + str(a2)
     answer = a1 * a2
  elif option == 4:
     problem = str(a1) + " // " + str(a2)
     answer = a1 // a2
  user_answer = get_user_input_solution(problem)
  res = solution_checking(user_answer, answer, count)
  return res

def get_user_input_solution(problem):
  print("Enter your answer")
  print(problem, end="")
  result = int(input(" = "))
  return result

def show_result(total, correct):
  print(f"You answered {total} questions with {correct} correct.")
  print(f"Your score is {round(correct*100/total,2)}%. Thank you.")

def solution_checking(user_solution, solution, count):
    if int(user_solution) == int(solution):
       res = count + 1
       print("Correct.")
    else:
       res = count
       print("Wrong.")
    return res


def main():
  total = 0
  correct = 0
  quiz_introduction()
  quiz_questions_init()
  section_separator()
  option = get_user_question_choice()
  while option != 5:
     total = total + 1
     correct = quiz_questions(option, correct)
     option = get_user_question_choice()
  print("Exit the quiz.")
  section_separator()
  show_result(total, correct)


main()
```
