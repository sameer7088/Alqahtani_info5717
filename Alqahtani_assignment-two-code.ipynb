```python
import os
import csv
from itertools import zip_longest

col1 = []
col2 = []
col3 = []
a = 0
for ss in range(1,100):
    path = '/Users/Sameer/Downloads/Assignment 2 data collection/ARTHROTEC.{}.txt'.format(ss)
    with open(path, 'r') as content_file:
        content = content_file.read()
    for data in content.split('\n'):
        col1.append('ARTHROTEC.{}'.format(ss) + '_'+str(a)+'.txt')
        col2.append(data)
        col3.append(''.join(m for m in data if m.isalnum()))
        a += 1

file = [col1, col2, col3]

export_data = zip_longest(*file, fillvalue = '')
with open('result.csv', 'w', encoding="ISO-8859-1", newline='') as myfile:
      wr = csv.writer(myfile)
      wr.writerow(("Sentence ID", "text content", "Without Punctuation"))
      wr.writerows(export_data)
myfile.close()
```
